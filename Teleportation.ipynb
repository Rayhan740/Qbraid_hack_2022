```python
from qiskit import *
```

```python
qc = QuantumCircuit(3,3)
#qc.x(0)
qc.barrier()
qc.draw("mpl")
```

```python
qc.h(1)
qc.cx(1,2)
qc.cx(0,1)
qc.h(0)
qc.barrier()
qc.draw()
```

```python
qc.measure([0,1],[0,1])
qc.barrier()
qc.draw()
```

```python
qc.cx(1,2)
qc.cz(0,2)
qc.draw(output = 'mpl')
```

```python
qc.measure(2,2)
simulator = Aer.get_backend("aer_simulator")
result = execute(qc, backend=simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
