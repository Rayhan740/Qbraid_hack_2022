## Basic Linear Algebra

As we represent bits in classical computer as 0 and 1. We represents our Quantum bit or qubit as the vector representation of $|0 \rangle$ and $|1 \rangle$. Which are vertical list of numbers:  $$ |0 \rangle =  \begin{bmatrix} 1 \\ 0 \end{bmatrix} , |1 \rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$   

A generic qubit with amplitudes ${\alpha}$ and ${\beta}$ would be:  $$ \begin{align} |{\psi} \rangle = {\alpha}|0 \rangle + {\beta}|1 \rangle 
 = {\alpha} \begin{bmatrix} 1 \\ 0 \end{bmatrix} + {\beta} \begin{bmatrix} 0 \\ 1 \end{bmatrix} 
 = \begin{bmatrix} {\alpha} \\ {\beta} \end{bmatrix} \end{align}$$ 

we can't represent any vector as quantum state there are some rule that need to be followed and one of them is the vector state need to be normalized that means $$ {\alpha}^2 + {\beta}^2 = 1 $$ 

One types of product we need to know for our computation is outer product. Suppose we have two $|0 \rangle $ state so there outer product will be $$ |0 \rangle |0 \rangle = |0 \rangle {\otimes} |0 \rangle = \begin{bmatrix} 1 \begin{bmatrix} 1 \\ 0 \end{bmatrix} \\ 0 \begin{bmatrix} 1 \\ 0 \end{bmatrix} \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}  = |00 \rangle $$

now we are ready for going into computation where we are going to learn how we can create an entangle state with multiple qubit. We know that there exist quantum states that can not be factored into product state. For example, with two qubits  , $$ \begin{align} |{\phi}^+ \rangle = \frac{1}{\sqrt{2}} (|00 \rangle + |11 \rangle) \end{align} $$


Which cannot be written as $ |{\psi_{1}} \rangle |{\psi_{0}} \rangle$. As a proof let's try to write it as a product state using the procedure from the Linear Algebra section: 
$$ |{\psi_{1}} \rangle |{\psi_{0}} \rangle  = (\alpha_{1}|0 \rangle + \beta_{1}|1 \rangle)(\alpha_{0}|0 \rangle + \beta_{0}|1 \rangle) =  {\alpha_{1}}{\alpha_{0}}|00 \rangle +  {\alpha_{1}}{\beta_{0}}|01 \rangle +  {\beta_{1}}{\alpha_{0}}|10 \rangle +  {\beta_{1}}{\beta_{0}}|11 \rangle $$


Matching the coefficient , we get $$ {\alpha_{1}}{\alpha_{0}} = {\frac{1}{\sqrt{2}}},\text{ } {\alpha_{1}}{\beta_{0}} = 0,\text{ } {\beta_{1}}{\alpha_{0}} = 0, {\beta_{1}}{\beta_{0}} =  {\frac{1}{\sqrt{2}}} $$

The Second equation requires $\alpha_{1}  = 0$ or $\beta_{0} = 0$. If $ \alpha_{1} = 0$, then the first equation gives $0 = \frac{1}{\sqrt{2}}$, which is false. If $ \beta_{0} = 0 $, then the fourth equation gives   $0 = \frac{1}{\sqrt{2}}$. Thus, there is no solution to these four equations, so $  |{\phi}^+ \rangle $ cannot be written as a product state. It is an entangled state. <b>This property that the state of the qubits are interwined is called entanglement</b>. Since an entangle state cannot be factored, a general entangled state of n qubits whould have $N = 2^n$ amplitueds $c_{0}$ through $c_{N-1}$: $$ |{\psi} \rangle = \sum \limits_{j=0}^{N-1} c_{j}|j \rangle = c_{0}|0 \rangle + c_{1} |1 \rangle +  .... + c_{N-1}|N-1 \rangle = \begin{bmatrix} c_{0} \\ c_{1} \\ ... \\ c_{N-1} \end{bmatrix} $$  

We know for two qubit the circuit that will give us an entangled state is: 

```python
# For two qubits 
from qiskit import *

qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.draw()
```

In the upper circuit we have seen that we used a <b> Hadamard gate </b> and a <b> CNOT gate </b> where the control is in 0 qubit and the target is in 1. The reason for using the Hadamard gate  we will be getting superposition state of two qubits like: $$ \frac{1}{\sqrt{2}} (|00 \rangle + |01 \rangle) $$ and after that when we are going to use CNOT gate then if the control is 1 then the target will be 1 as well. So, the final state will be: $$ \frac{1}{\sqrt{2}} (|00 \rangle + |11 \rangle) $$

So, I guess we get the idea of how we can create an entangled state so let's create an entangled state with multiple qubit.

## Creating entanglement with multiple qubits

```python
from qiskit import * # From qiskit import everything

nqubits = 3 # Taking 3 qubits
nclassical = 3 # Taking 3 classical bits
circuits = QuantumCircuit(nqubits,nclassical) # define a QuantumCircuit with 3 Qubit and 3 Classical bit
circuits.h(0)  # Using ane hadamard gate to create a superposition state
circuits.cx(0,1) # Using two CNOT gate so that we can create a Bell state or entangled state
circuits.cx(1,2) 
circuits.measure([0,1,2],[0,1,2]) # Measuring the circuit
```

```python
# Here we can see the circuit representation of our code above
%matplotlib inline
circuits.draw(output='mpl') 
```

From the circuit above we can see that we have used one hadamard gate and two CNOT gate. Let's check the statevector that we are going to get from this circuit:

```python
from qiskit.quantum_info import Statevector

nqubits = 3
nclassical = 3
circuits = QuantumCircuit(nqubits,nclassical)
circuits.h(0)
circuits.cx(0,1)
circuits.cx(1,2)

ket = Statevector(circuits)
ket.draw(output='latex')
```

From the statevector we can say that we cannot get the product state of $|{\psi_{0}} \rangle |{\psi_{1}} \rangle |{\psi_{2}} \rangle$. So, we can surely say that the statevector that is being shown in above is an entangled state. Now let's see verify our answer in local simulator and then plot the histogram $:$

```python
from qiskit.visualization import plot_histogram 


nqubits = 3
nclassical = 3
circuits = QuantumCircuit(nqubits,nclassical)
circuits.h(0)
circuits.cx(0,1)
circuits.cx(1,2)
circuits.measure([0,1,2],[0,1,2]) 

simulator = Aer.get_backend('qasm_simulator') # Taking qasm_simulator as local simulator
sim_result = execute(circuits, backend=simulator, shots =1024).result() # Executing the circuit and getting the result
plot_histogram(sim_result.get_counts(circuits))
```

As a simple model of noise in this process, we can imagine that the measurement first selects one of these outputs in a perfect and noiseless manner, and then noise subsequently causes this perfect output to be randomly pertubed before it is return to the user.
